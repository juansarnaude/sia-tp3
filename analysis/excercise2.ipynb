```python
import matplotlib.pyplot as plt
import pandas as pd

def read_csv_data(file_path):
    data = pd.read_csv(file_path)
    return data['period'].tolist(), data['error'].tolist()

epochs_1, errors_1 = read_csv_data('../output/ej2/output_linear.csv')
epochs_2, errors_2 = read_csv_data('../output/ej2/output_nonlinear_sigmoid.csv')
epochs_3, errors_3 = read_csv_data('../output/ej2/output_nonlinear_tanh.csv')

plt.figure(figsize=(10, 6))

plt.plot(epochs_1, errors_1, label='Linear')  
plt.plot(epochs_2, errors_2, label='Non Linear (Sigmoid)')
plt.plot(epochs_3, errors_3, label='Non Linear (Tanh)')

plt.title('Evolution of Errors over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error Value')
plt.legend(title='Errors')

plt.grid(True)
plt.show()
```
