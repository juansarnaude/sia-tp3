```python
import plotly.graph_objects as go
import pandas as pd

def read_csv_data(file_path):
    data = pd.read_csv(file_path)
    return data['period'].tolist(), data['error'].tolist()

epochs_1, errors_1 = read_csv_data('../output/ej2/output_linear.csv')
epochs_2, errors_2 = read_csv_data('../output/ej2/output_nonlinear_sigmoid.csv')
epochs_3, errors_3 = read_csv_data('../output/ej2/output_nonlinear_tanh.csv')

fig = go.Figure()

fig.add_trace(go.Scatter(x=epochs_1, y=errors_1, mode='lines+markers', name='Linear'))
fig.add_trace(go.Scatter(x=epochs_2, y=errors_2, mode='lines+markers', name='Non Linear (Sigmoid)'))
fig.add_trace(go.Scatter(x=epochs_3, y=errors_3, mode='lines+markers', name='Non Linear (Tanh)'))

fig.update_layout(
    title='Evolution of Errors over Epochs',
    xaxis_title='Epochs',
    yaxis_title='Error Value',
    legend_title='Errors',
    yaxis=dict(type='log'),  
    template='plotly',
)


fig.show()
```
