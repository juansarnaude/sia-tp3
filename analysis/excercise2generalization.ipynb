```python
import numpy as np
# ERROR FROM BETAS OVER TIME
# Need a file with 10 betas in outputs/tanh/betatest 

import plotly.graph_objects as go
import pandas as pd

def read_csv_data(file_path):
    data = pd.read_csv(file_path)
    return data['period'].tolist(), data['testing_error'].tolist(), data['training_error'].tolist()

nums = [0,1,2,3]
strings = ["sigmoid", "tanh"]

for num in nums:
    #epochs_1, test_errors_1, train_errors_1 = read_csv_data('../output/ej2/output_nonlinear_tanh' + str(num) + '.csv') 
    epochs_2, test_errors_2, train_errors_2 = read_csv_data('../output/ej2/output_nonlinear_sigmoid' + str(num) + '.csv') 

    fig = go.Figure()
    
    #fig.add_trace(go.Scatter(x=epochs_1[:1500], y=test_errors_1[:1500], mode='lines', line=dict(color='blue'), name='Testing Error (Tanh)' ))
    #fig.add_trace(go.Scatter(x=epochs_1[:1500], y=train_errors_1[:1500], mode='lines', line=dict(dash='dot', color='blue'), name='Training Error (Sigmoid)'))
    
            
    fig.add_trace(go.Scatter(x=epochs_2[:1500], y=test_errors_2[:1500], mode='lines', line=dict(color='green'), name='Testing Error'))
    fig.add_trace(go.Scatter(x=epochs_2[:1500], y=train_errors_2[:1500], mode='lines', line=dict(color='blue'), name='Training Error'))
    
    
    
    fig.update_layout(
        title='Testing and Training MSE for different Non-Linear Perceptron k=4, Set nÂ°=' + str(num+1),
        xaxis_title='Epochs',
        yaxis_title='MSE',
        legend_title='Errors',
        yaxis=dict(type='log'),
        template='plotly'
    )
    
    fig.show()
    
test_errors_total = np.zeros(4999)
train_errors_total = np.zeros(4999)

for num in nums:
    epochs_2, test_errors_2, train_errors_2 = read_csv_data('../output/ej2/output_nonlinear_sigmoid' + str(num) + '.csv') 

            
    test_errors_total = np.array(test_errors_total) + np.array(test_errors_2)
    train_errors_total = np.array(train_errors_total) + np.array(train_errors_2)
    
    
fig = go.Figure()

fig.add_trace(go.Scatter(x=epochs_2[:1500], y=test_errors_total[:1500], mode='lines', line=dict(color='green'), name='Testing Error'))
fig.add_trace(go.Scatter(x=epochs_2[:1500], y=train_errors_total[:1500], mode='lines', line=dict(color='blue'), name='Training Error'))
    
fig.update_layout(
    title='Testing and Training MSE for different Non-Linear Perceptron k=4, Set totals',
    xaxis_title='Epochs',
    yaxis_title='MSE',
    legend_title='Errors',
    yaxis=dict(type='log'),
    template='plotly'
)

fig.show()

```
